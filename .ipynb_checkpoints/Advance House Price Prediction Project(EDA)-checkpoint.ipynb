#### Data Analysis Phase

```python
## Import necessary libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
## read the csv file
dataset = pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
## Print the shape of dataset with rows and columns
print(dataset.shape)  ## 1460 rows and 81 columns
```

```python
dataset.head(50)
```

### We will analyze:
1. Missing values
2. All the numerical variables
3. Distribution of the Numerical Variables
4. Categorical variables
5. Cardiinality of Categorical Variables
6. Outliers
7. Relationship between independent and dependent feature(SalePrice)

```python
### 1. Missing Values
feature_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
feature_with_na
```

```python
for feature in feature_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4), '% missing values')
```

#### since there are many missing values we need to find the relationship between missing vlaues and sales price

```python
for feature in feature_with_na:
    data = dataset.copy()

    data[feature] = np.where(data[feature].isnull(),1,0) 

    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### With relation between the missing values and the dependent variable is clearly visible 

### So need to replce these nan values with something meaningful which we will do in the feature engineering

#### Numerical Variables

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
numerical_features
```

```python
print(f'Number of numerical variables: {len(numerical_features)}')
```

```python
## Visualize the numerical variable 
dataset[numerical_features].head()
```

```python
## List of variables that contain year
year_feature = [feature for feature in numerical_features if 'Year' in feature or 'Yr' in feature]
```

```python
year_feature
```

```python
## getting to know the unique values these year_feature contains
for feature in year_feature:
    print(feature,' : ', dataset[feature].unique())
```

#### Analyze the temporal Datetime

```python
dataset.groupby('YearBuilt')['SalePrice'].median().plot()
plt.xlabel('Year BUilt')
plt.ylabel('Median Hourse Price')
plt.title('House Price vs Year Built')
plt.show()
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median Hourse Price')
plt.title('House Price vs Year Sold')
plt.show()
```

```python
### We will comprae the difference between all years feature with saleprice

for feature in year_feature:
    data = dataset.copy()
    # print(data['YrSold'].head(15))
    # print(data[feature].head(15))
    # print((data['YrSold'] - data[feature]).head(15))
    data[feature] = data['YrSold'] - data[feature]
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
```

### Numerical Features i.e. Discrete Feature and Continuous Feature

```python
numerical_features
```

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())>25 and feature not in year_feature+['Id']]
discrete_feature
print(f'Discrete Features Count {len(discrete_feature)}')
```

#### Showing the relationship between the discrete feature and SalePrice using bar plot

```python
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(f'Sale Price vs {feature}')
    plt.show()
```

```python
### 3. Continuous Variable
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
```

```python
continuous_feature
```

```python
print(f'Continuous Features Count {len(continuous_feature)}')
```

### Showing the relationship between the continous feature and Sale Price

```python
for feature in continuous_feature:
    data = dataset.copy()
    print('-------------------------------')
    print(data[feature].head())
    print('-------------------------------')
```

```python
for feature in continuous_feature:
    data = dataset.copy()
    # data[feature].hist(bins=20)
    sns.histplot(data=data, x=feature, kde=True, bins=20,color='red')
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

#### Using Logarithmic expression

```python
continuous_feature
```

```python
for feature in continuous_feature:
    data = dataset.copy()

    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        # plt.scatter(data[feature],data['SalePrice'])
        sns.histplot(data=data, x=feature, kde=True, bins=20,color="red")
        plt.xlabel(feature)
        plt.ylabel("Sale Price")
        plt.title(feature)
        plt.show()
```

### Outlier

```python
for feature in continuous_feature:
    data = dataset.copy()

    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

### Categorical Feature 

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
categorical_features
```
